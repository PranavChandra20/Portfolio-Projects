```python
#import libraries

import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib 
import matplotlib.pyplot as plt
plt.style.use('ggplot')

from matplotlib.pyplot import figure

matplotlib.rcParams['figure.figsize'] = (12,8)

#read in the data

movies = pd.read_csv(r'C:\Users\prana\OneDrive\Documents\AlexTheAnalyst\python correlation\movies.csv')
```

```python
#lets look at the data 

movies.head()
```

```python
# checking if there is any missing data

movies = movies.dropna()

for col in movies.columns :
    pct_missing = np.mean(movies[col].isnull())
    print(col, pct_missing)
    
```

```python
#checking the data types
movies.dtypes
```

```python
#chaneg datatypes of columns

movies['budget'] = movies['budget'].astype('int64')
movies['gross'] = movies['gross'].astype('int64')
```

```python
movies.head()
```

```python
#create correct year column

movies['CorrectYear'] = movies['released'].astype(str).str[:4]

movies.head()
```

```python
movies.sort_values(by=['gross'], inplace = False, ascending = False)
```

```python
pd.set_option('display.max_rows', None)
```

```python
movies.sort_values(by=['gross'], inplace = False, ascending = False)
movies.head()
```

```python
movies.drop_duplicates()
movies.head()
```

```python
#scatter plot to compare budget and gross


plt.scatter(x=movies['budget'],y=movies['gross'])

plt.title('Budget vs Gross Revenue')
plt.xlabel('Budget')
plt.ylabel('Gross')
plt.show()
```

```python
# budget vs gross using seaborn

sns.regplot(x='budget', y='gross', data=movies, scatter_kws = {"color":"black"}, line_kws={"color":"green"})
```

```python
#looking at correlations

movies.corr()
```

```python
correlation_matrix = movies.corr()

sns.heatmap(correlation_matrix, annot = True)
plt.title("Correlation Matrix")
plt.xlabel("Movie features")
plt.ylabel("Movie features")
plt.show()
```

```python
movies.head()
```

```python
movies_numerized = movies

for col_name in movies_numerized.columns:
    if(movies_numerized[col_name].dtype == 'object'):
        movies_numerized[col_name] = movies_numerized[col_name].astype('category')
        movies_numerized[col_name] = movies_numerized[col_name].cat.codes
        
movies_numerized.head()
```

```python
correlation_matrix = movies_numerized.corr()

sns.heatmap(correlation_matrix, annot = True)
plt.title("Correlation Matrix")
plt.xlabel("Movie features")
plt.ylabel("Movie features")
plt.show()
```

```python
movies_numerized.corr()
```

```python
correlation_mat = movies_numerized.corr()

corr_pairs = correlation_mat.unstack()

corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()

sorted_pairs
```

```python
high_correlation = sorted_pairs[(sorted_pairs) > 0.5 ]

high_correlation
```
